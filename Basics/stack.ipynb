### Valid Parentheses
##### 리트코드 20 (https://leetcode.com/problems/valid-parentheses/description/)

#### 1. (),[],{} 다음과같은 세종류의 괄호를 검사하는 프로그램을 만들어라

#### 알고리즘
##### 1. (,[,{ 가 들어올시 스택에 넣는다
##### 2. ),],} 가 들어오면 스택에 꺼내서 짝을 맞춘후 제거한다.
##### 3. 마지막에 스택을 검사하여 모든 문자열을 사용되었는지 확인한다.
#### 입력 데이터
##### s = "()[]{}"
#### 출력 데이터
##### True

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        index = 0
        while index != len(s):
            cur = s[index]
            if cur == ')':
                if not stack:
                    return False
                if stack.pop() !='(':
                    return False
            elif cur == ']':
                if not stack:
                    return False
                if stack.pop() !='[':
                    return False
            elif cur == '}':
                if not stack:
                    return False
                if stack.pop() !='{':
                    return False
            else:
                stack.append(cur)
            index+=1
        if stack:
            return False
        else:
            return True
```

### Daily temperature
##### 리트코드 (https://leetcode.com/problems/daily-temperatures/description/)

#### 1. 각 일일 기온을 기록한 리스트가 있다. 오늘 기온에서 기온이 증가할려면 몇일이 걸릴지 찾아라.

#### 알고리즘
##### 1. 스택 차례로 기온을 자신의 인덱스와함께 입력한다.
##### 2. 스택이 쌓여있다면 맨 아래 요소가 가장 기온이 높은 것이다.
##### 3. 매 차례에 스택의 가장 윗부분의 요소와 비교 후 윗부분 요소의 기온이 작다면 정답 리스트에 (현재 요소인덱스 - 스텍 윗부분 요소 인덱스)를 하여 기록한다.
#### 입력 데이터
##### temperatures = [73,74,75,71,69,72,76,73]
#### 출력 데이터
##### Output: [1,1,4,2,1,1,0,0]

```python
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        stack = []
        answer = [0]*len(temperatures)
        for i, v in enumerate(temperatures):
            while stack:
                cur_v, cur_i=stack.pop()
                if cur_v < v:
                    answer[cur_i] = i-cur_i
                else:
                    stack.append((cur_v,cur_i))
                    break
            stack.append((v,i))

        return answer

```

### 주식 가격
##### 프로그래머 (https://school.programmers.co.kr/learn/courses/30/lessons/42584)

#### 1. 일일 주식가격을 기록한 리스트가 있다. 해당 주식이 내려가지 않은 일수를 계산하라.

#### 알고리즘
##### 1. 스택 차례로 주식 가격을 자신의 인덱스와함께 입력한다.
##### 2. 매 차례에 스택의 가장 윗부분의 요소와 비교 후 윗부분 요소의 주식이 더 크다면 정답 리스트에 (현재 요소인덱스 - 스텍 윗부분 요소 인덱스)를 하여 기록한다.
##### 3. 스택이 쌓여있다면 맨 아래 요소가 가장 주식이 가장 낮은 것이다.
##### 4. 반복이 모두 끝난다면 요일이 0인 요소는 (마지막 인덱스 - 현재 인덱스)로 처리하여 처리한다.
#### 입력 데이터
##### [1, 2, 3, 2, 3]
#### 출력 데이터
##### [4, 3, 1, 1, 0]

```python
def solution(prices):
    stack = []
    answer = [0] *len(prices)
    for i, v in enumerate(prices):
        while stack:
            cur_i,cur_v = stack.pop()
            if cur_v > v:
                answer[cur_i] = i - cur_i
            else:
                stack.append((cur_i,cur_v))
                break
        stack.append((i,v))

    for i in range(len(answer)):
        if answer[i] == 0:
            answer[i] = len(answer)-1-i

    return answer
```
