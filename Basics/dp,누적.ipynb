### Climbing Stairs(dp)
##### 리트코드(https://leetcode.com/problems/climbing-stairs/description/)

#### 1. n 개의 계단을 올라야 목적지에 다다를 수 있고 한번에 1계단 혹은 2계단씩 오를수 있다.
#### 2. 오를수 있는 방법의개수를 구하자

#### 알고리즘
##### 1. 계단을 한번에 1번 또는 2번오를 수있기때문에 1칸 계단을 오르는 방법은 1step으로 하나이고 2칸 계단을 오르는 방법은 1step+1step or 2step으로 2가지 방법이 있다.
##### 2. 3칸 계단을 오르는 방법은 1칸 계단 오르는 경우의수와 2칸 계단 오르는 경우의 수를 더한값과 같다. 왜냐하면 2칸오른시점에서 3칸을 갈려면 1칸을 더 올라야 되기때문이다.
##### 3. 해당 논리를 바탕으로 수식을 세우면 n칸 = n-1칸 + n-2칸으로 되어진다.
##### 4. 세운 점화식을 DP방식으로 코딩하면된다.

#### 입력 데이터
##### n = 4
##### m = 4
##### [[2, 3], [3, 3]]
#### 출력 데이터
##### 5

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        if n < 2:
            return 1
        list_ = [0]*(n+1)
        list_[1] = 1
        list_[2] = 2
        for i in range(3, len(list_)):
            list_[i] = list_[i-2]+list_[i-1]
        return list_[n]
```

### 피보나치 (dp,완전탐색)
##### 리트코드(https://school.programmers.co.kr/learn/courses/30/lessons/12945)

#### 1. 피보나치 수는 F(0) = 0, F(1) = 1일 때, 1 이상의 n에 대하여 F(n) = F(n-1) + F(n-2) 가 적용되는 수 이다
#### F(n)을 구하고 그 값을 1234567으로 나눈 나머지를 리턴하라

#### 알고리즘

#### 입력 데이터
#### 출력 데이터

```python
def solution(n):
    list_  = [0]*(n+1)
    list_[1] = 1
    for i in range(2,n+1):
        list_[i] = list_[i-1]+list_[i-2]
    return list_[::-1][0]%1234567
```
